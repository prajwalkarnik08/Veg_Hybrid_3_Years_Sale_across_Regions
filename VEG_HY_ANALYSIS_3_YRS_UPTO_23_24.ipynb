# Cowpea

### BUBBLY

```python
df_bubbly = pd.DataFrame([])
df_bubbly["NSQ of BUBBLY 21-22"] = df[(df['PRODUCT NAME'] == "BUBBLY")].groupby(df["REGION"])["NET SALE QTY 21-22"].sum()
df_bubbly["NSA of BUBBLY 21-22"] = df[(df['PRODUCT NAME'] == "BUBBLY")].groupby(df["REGION"])["NET SALE AMT 21-22"].sum()
df_bubbly["NSQ of BUBBLY 22-23"] = df[(df['PRODUCT NAME'] == "BUBBLY")].groupby(df["REGION"])["NET SALE QTY 22-23"].sum()
df_bubbly["NSA of BUBBLY 22-23"] = df[(df['PRODUCT NAME'] == "BUBBLY")].groupby(df["REGION"])["NET SALE AMT 22-23"].sum()
df_bubbly["NSQ of BUBBLY 23-24"] = df[(df['PRODUCT NAME'] == "BUBBLY")].groupby(df["REGION"])["NET SALE QTY 23-24"].sum()
df_bubbly["NSA of BUBBLY 23-24"] = df[(df['PRODUCT NAME'] == "BUBBLY")].groupby(df["REGION"])["NET SALE AMT 23-24"].sum()
df_bubbly
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_zhakas7 is already defined and contains the data
# Reset the index to make 'REGION' a column
#df_bubbly = df_bubbly.reset_index()

# Define regions (x-axis labels)
regions = df_bubbly["REGION"]

# Define the years and corresponding column names for plotting
years = ["21-22", "22-23", "23-24"]
#nsq_columns = [f"NSQ of BUBBLY {year}" for year in years]
nsa_columns = [f"NSA of BUBBLY {year}" for year in years]

# Set the positions of the bars
x = np.arange(len(regions))  # x positions for the bars
bar_width = 0.25  # Width of each bar

# Create the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Plot NSQ data (Net Sale Quantity) for different years
#for i, year in enumerate(years):
#    ax.bar(x + i * bar_width, df_bubbly[nsq_columns[i]], bar_width, label=f"NSQ {year}")

# Plot NSA data (Net Sale Amount) for different years
for i, year in enumerate(years):
    ax.bar(x + (i + len(years)) * bar_width, df_bubbly[nsa_columns[i]], bar_width, label=f"NSA {year}", alpha=0.7)

# Add labels, title, and other plot details
ax.set_xlabel("Regions")
ax.set_ylabel("Values")
#ax.set_title("Net Sale Quantity (NSQ) and Net Sale Amount (NSA) of BUBBLY Across Regions")
ax.set_title("Net Sale Amount (NSA) of BUBBLY Across Regions")
ax.set_xticks(x + bar_width * (len(years)))  # Adjust the x-ticks to center the bars
ax.set_xticklabels(regions, rotation=100)
ax.legend(title="Year", loc="upper left")

# Add gridlines for better readability
ax.grid(axis="y", linestyle="--", alpha=0.7)

# Adjust layout to avoid overlap and display the plot
plt.tight_layout()
plt.show()
```
